```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
df_x = pd.read_csv('iPhone Thing-Accelerometer_X.csv')
df_y = pd.read_csv('iPhone Thing-Accelerometer_Y.csv')
df_z = pd.read_csv('iPhone Thing-Accelerometer_Z.csv')

# Plot Accelerometer X Data
plt.figure(figsize=(10, 6))
plt.plot(df_x['time'], df_x['value'], label='Accelerometer X')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Accelerometer X Data')
plt.legend()
plt.show()

# Plot Accelerometer Y Data
plt.figure(figsize=(10, 6))
plt.plot(df_y['time'], df_y['value'], label='Accelerometer Y', color='green')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Accelerometer Y Data')
plt.legend()
plt.show()

# Plot Accelerometer Z Data
plt.figure(figsize=(10, 6))
plt.plot(df_z['time'], df_z['value'], label='Accelerometer Z', color='red')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Accelerometer Z Data')
plt.legend()
plt.show()

# Plot All Three Variables Together
plt.figure(figsize=(10, 6))
plt.plot(df_x['time'], df_x['value'], label='Accelerometer X')
plt.plot(df_y['time'], df_y['value'], label='Accelerometer Y', color='green')
plt.plot(df_z['time'], df_z['value'], label='Accelerometer Z', color='red')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Accelerometer X, Y, Z Data')
plt.legend()
plt.show()
```
